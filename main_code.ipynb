```python
x = torch.FloatTensor(x)
y = torch.FloatTensor(y)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 32)
```

```python
import torch
import torch.nn as nn

# تعریف معماری ResNet با Sequential
model = nn.Sequential(
    # لایه تبدیل تعداد کانال‌ها به ۳
    nn.Conv2d(1, 3, kernel_size=1, bias=False),  # تبدیل داده‌های تک‌کاناله به ۳ کانال
    
    # لایه ابتدایی
    nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False),
    nn.BatchNorm2d(64),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=3, stride=2, padding=1),
    
    # بلوک‌های ResNet ساده‌شده
    nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.ReLU(inplace=True),
    nn.Conv2d(128, 256, kernel_size=3, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.ReLU(inplace=True),
    nn.Conv2d(256, 512, kernel_size=3, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(512),
    nn.ReLU(inplace=True),
    
    # لایه‌های پایانی
    nn.AdaptiveAvgPool2d((1, 1)),
    nn.Flatten(),
    nn.Linear(512, 7)  # تغییر تعداد نودهای خروجی به 7 برای مسئله رگرسیون
)

# نمایش معماری مدل
print(model)
```

```python
# تعریف معماری ResNet با لایه‌های بیشتر
model = nn.Sequential(
    # لایه تبدیل تعداد کانال‌ها به ۳
    nn.Conv2d(1, 3, kernel_size=1, bias=False),  # تبدیل داده‌های تک‌کاناله به ۳ کانال
    
    # لایه ابتدایی
    nn.Conv2d(3, 64, kernel_size=7, stride=2, padding=3, bias=False),
    nn.BatchNorm2d(64),
    nn.ReLU(inplace=True),
    nn.MaxPool2d(kernel_size=3, stride=2, padding=1),
    
    # بلوک 1
    nn.Conv2d(64, 64, kernel_size=3, stride=1, padding=1, bias=False),
    nn.BatchNorm2d(64),
    nn.ReLU(inplace=True),
    nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.ReLU(inplace=True),

    # بلوک 2
    nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1, bias=False),
    nn.BatchNorm2d(128),
    nn.ReLU(inplace=True),
    nn.Conv2d(128, 256, kernel_size=3, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.ReLU(inplace=True),

    # بلوک 3
    nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1, bias=False),
    nn.BatchNorm2d(256),
    nn.ReLU(inplace=True),
    nn.Conv2d(256, 512, kernel_size=3, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(512),
    nn.ReLU(inplace=True),

    # بلوک 4
    nn.Conv2d(512, 512, kernel_size=3, stride=1, padding=1, bias=False),
    nn.BatchNorm2d(512),
    nn.ReLU(inplace=True),
    nn.Conv2d(512, 1024, kernel_size=3, stride=2, padding=1, bias=False),
    nn.BatchNorm2d(1024),
    nn.ReLU(inplace=True),

    # لایه‌های پایانی
    nn.AdaptiveAvgPool2d((1, 1)),
    nn.Flatten(),
    nn.Linear(1024, 7)  # تعداد خروجی‌ها 7 برای مسئله رگرسیون
)

# نمایش معماری مدل
print(model)
```

# Train & Test

```python
# آموزش مدل
num_epochs = 550
loss_train_hist, loss_test_hist = [], []

for epoch in range(num_epochs):
    model.train()  # حالت آموزش
    train_loss = 0

    for x_batch, y_batch in train_loader:
        # بررسی و تنظیم ابعاد داده
        if x_batch.dim() == 2:  # اگر داده‌ها به صورت 2D (batch_size, length) هستند
            x_batch = x_batch.unsqueeze(1).unsqueeze(1)  # افزودن بعد کانال و ارتفاع/عرض
        elif x_batch.dim() == 3:  # اگر داده‌ها به صورت 3D (batch_size, channels, length) هستند
            x_batch = x_batch.unsqueeze(1)  # افزودن بعد کانال برای ارتفاع/عرض
        elif x_batch.shape[1] != 3:  # اگر تعداد کانال‌ها 3 نیست
            x_batch = x_batch.expand(-1, 3, -1, -1)  # تبدیل به ۳ کانال
        
        # انجام پیش‌بینی
        y_pre = model(x_batch)
        
        # محاسبه Loss
        loss = loss_fn(y_pre, y_batch)
        
        # به‌روزرسانی وزن‌ها
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        train_loss += loss.item()
    
    avg_train_loss = train_loss / len(train_loader)
    loss_train_hist.append(avg_train_loss)
    print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {avg_train_loss:.4f}")
    
    # ارزیابی مدل
    model.eval()  # حالت ارزیابی
    test_loss = 0
    with torch.no_grad():
        for x_batch, y_batch in test_loader:
            # تنظیم ابعاد داده برای ارزیابی
            if x_batch.dim() == 2:
                x_batch = x_batch.unsqueeze(1).unsqueeze(1)
            elif x_batch.dim() == 3:
                x_batch = x_batch.unsqueeze(1)
            elif x_batch.shape[1] != 3:
                x_batch = x_batch.expand(-1, 3, -1, -1)
            
            y_pre = model(x_batch)
            loss = loss_fn(y_pre, y_batch)
            test_loss += loss.item()
    
    avg_test_loss = test_loss / len(test_loader)
    loss_test_hist.append(avg_test_loss)
    print(f"Epoch [{epoch+1}/{num_epochs}], Test Loss: {avg_test_loss:.4f}")
    print('-' * 60)

# نمایش Loss نهایی
print(f"Final Test Loss: {avg_test_loss:.4f}")
```
